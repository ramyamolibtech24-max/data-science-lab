Scratch implementation

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load digits dataset
digits = load_digits()
X, y = digits.data, digits.target

# Standardize the data (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# PCA implementation from scratch
def compute_pca(X, n_components):
    print("welcome to PCA from scratch")

    # Step 1ï¸âƒ£: Compute covariance matrix
    covariance_matrix = np.cov(X.T)

    # Step 2ï¸âƒ£: Compute eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix)

    # Step 3ï¸âƒ£: Sort eigenvalues and eigenvectors in descending order
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]

    # Step 4ï¸âƒ£: Select top 'n_components' eigenvectors
    components = eigenvectors[:, :n_components]

    # Step 5ï¸âƒ£: Transform the data to new subspace
    X_reduced = np.dot(X, components)

    # Step 6ï¸âƒ£: Calculate explained variance ratio
    explained_variance_ratio = eigenvalues[:n_components] / np.sum(eigenvalues)

    return X_reduced, components, explained_variance_ratio

# ------------------------------------------------------------
# Apply PCA
# ------------------------------------------------------------
X_pca, components, evr = compute_pca(X_scaled, n_components=2)
print("\nExplained variance ratio for first 2 components:", evr)
print("Total variance captured:", np.sum(evr))

# ------------------------------------------------------------
# Visualize 2D PCA projection
# ------------------------------------------------------------
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab10', s=15)
plt.legend(*scatter.legend_elements(), title="Digits", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Digits Dataset projected onto 2D PCA space")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# Optional: Train-test split and Logistic Regression using PCA-reduced data
# ------------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy using 2 PCA components:", accuracy_score(y_test, y_pred))



```

```python


# Reduce to 30 principal components
n_components = 30
X_pca, components, eigenvalues = compute_pca(X_scaled, n_components)

# Split data into train/test
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Train a classifier (Logistic Regression)
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

# Predict and evaluate
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Classification accuracy after PCA: {accuracy:.4f}")

# Optional: plot the first 2 principal components
plt.figure(figsize=(8,6))
for digit in np.unique(y):
    plt.scatter(X_pca[y==digit, 0], X_pca[y==digit, 1], label=str(digit), alpha=0.6)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Digits Dataset Projected onto First 2 Principal Components')
plt.legend()
plt.show()
```

We get less accuancy (~60%) as using only 2 components did not retain much of the feature information. However in real life you will find many cases where using 2 or few PCA components can still give you a pretty good accuracy

Download heart disease dataset heart.csv in Exercise folder and do following, (credits of dataset: https://www.kaggle.com/fedesoriano/heart-failure-prediction)

Load heart disease dataset in pandas dataframe
Remove outliers using Z score. Usual guideline is to remove anything that has Z score > 3 formula or Z score < -3
Convert text columns to numbers using label encoding and one hot encoding
Apply scaling
Build a classification model using various methods (SVM, logistic regression, random forest) and check which model gives you the best accuracy
Now use PCA to reduce dimensions, retrain your model and see what impact it has on your model in terms of accuracy. Keep in mind that many times doing PCA reduces the accuracy but computation is much lighter and that's the trade off you need to consider while building models in real life

```python
# Full pipeline for heart.csv
# Put heart.csv in /mnt/data/heart.csv or change file_path below.
import os, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
from scipy import stats
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"C:\Users\Ramya Oli\Downloads\heart.csv"  # <- change if needed

if not os.path.exists(file_path):
    raise FileNotFoundError(f"Could not find '{file_path}'. Upload heart.csv to /mnt/data or change file_path.")

# 1) Load dataset
df = pd.read_csv(file_path)
print("Loaded shape:", df.shape)
display(df.head())

# 2) Remove outliers using Z-score (apply to numeric columns)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print("Numeric columns:", numeric_cols)

# calculate z-scores and filter
z = np.abs(stats.zscore(df[numeric_cols], nan_policy='omit'))
z = np.nan_to_num(z)  # in case some columns are constant
rows_keep = (z < 3).all(axis=1)
df_no_out = df.loc[rows_keep].reset_index(drop=True)
print(f"Rows before: {df.shape[0]} | after Z<3 filtering: {df_no_out.shape[0]} (removed {df.shape[0]-df_no_out.shape[0]})\n")

# 3) Encode text columns -> detect object/category columns
cat_cols = df_no_out.select_dtypes(include=['object', 'category']).columns.tolist()
print("Categorical columns:", cat_cols)

df_enc = df_no_out.copy()

for col in cat_cols:
    if df_enc[col].nunique() == 2:
        le = LabelEncoder()
        df_enc[col] = le.fit_transform(df_enc[col])
        print(f"Label-encoded {col}")
    else:
        dummies = pd.get_dummies(df_enc[col], prefix=col, drop_first=True)
        df_enc = pd.concat([df_enc.drop(columns=[col]), dummies], axis=1)
        print(f"One-hot encoded {col} -> {dummies.shape[1]} cols")

print("\nShape after encoding:", df_enc.shape)

# 4) Identify target column (common names)
possible_targets = ['target','Target','DEATH_EVENT','death_event','output','HeartDisease','heart_disease','cardio','disease']
target_col = None
for t in possible_targets:
    if t in df_enc.columns:
        target_col = t
        break

# fallback heuristics: binary column with 'death' in name or last column
if target_col is None:
    for c in df_enc.columns:
        if df_enc[c].nunique() == 2 and ('death' in c.lower() or 'target' in c.lower() or 'event' in c.lower() or 'disease' in c.lower()):
            target_col = c
            break

if target_col is None:
    # fallback to last column (inform user)
    print("Could not auto-detect target column; falling back to last column as target.")
    target_col = df_enc.columns[-1]

print("Using target column:", target_col, "| unique values:", df_enc[target_col].unique())

X = df_enc.drop(columns=[target_col])
y = df_enc[target_col]

# 5) Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 6) Train/test split (stratified if possible)
try:
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
except:
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 7) Train models and evaluate
models = {
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'SVM_rbf': SVC(kernel='rbf', probability=True),
    'RandomForest': RandomForestClassifier(n_estimators=200, random_state=42)
}

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = {'accuracy': acc, 'report': classification_report(y_test, y_pred, zero_division=0), 'confusion': confusion_matrix(y_test, y_pred)}
    print(f"\n{name} accuracy: {acc:.4f}")
    print(results[name]['report'])

# 8) Apply PCA (retain 95% variance), retrain & evaluate
pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_scaled)
print(f"\nPCA reduced shape: {X_pca.shape} | components: {pca.n_components_} | explained var sum: {pca.explained_variance_ratio_.sum():.4f}")

# split PCA data
try:
    Xtr_p, Xte_p, ytr_p, yte_p = train_test_split(X_pca, y, test_size=0.2, random_state=42, stratify=y)
except:
    Xtr_p, Xte_p, ytr_p, yte_p = train_test_split(X_pca, y, test_size=0.2, random_state=42)

pca_results = {}
for name in models:
    # create fresh models
    if name == 'LogisticRegression':
        m = LogisticRegression(max_iter=1000)
    elif name == 'SVM_rbf':
        m = SVC(kernel='rbf', probability=True)
    else:
        m = RandomForestClassifier(n_estimators=200, random_state=42)
    m.fit(Xtr_p, ytr_p)
    yp = m.predict(Xte_p)
    accp = accuracy_score(yte_p, yp)
    pca_results[name] = {'accuracy': accp, 'report': classification_report(yte_p, yp, zero_division=0), 'confusion': confusion_matrix(yte_p, yp)}
    print(f"\n{name} accuracy after PCA: {accp:.4f}")
    print(pca_results[name]['report'])

# 9) Summary table
summary = pd.DataFrame([{
    'Model': name,
    'Acc_before_PCA': results[name]['accuracy'],
    'Acc_after_PCA': pca_results[name]['accuracy']
} for name in results])
display(summary)

# 10) Plot accuracies for quick visual
plt.figure(figsize=(8,4))
x = np.arange(len(summary))
plt.bar(x - 0.15, summary['Acc_before_PCA'], width=0.3, label='Before PCA')
plt.bar(x + 0.15, summary['Acc_after_PCA'], width=0.3, label='After PCA')
plt.xticks(x, summary['Model'])
plt.ylabel('Accuracy')
plt.title('Model Accuracies: Before vs After PCA')
plt.legend()
plt.ylim(0,1)
plt.show()

# 11) Save summary and PCA info
summary.to_csv('/mnt/data/heart_model_summary.csv', index=False)
pca_info = {'n_components': pca.n_components_, 'explained_variance': float(pca.explained_variance_ratio_.sum())}
print("Saved summary to /mnt/data/heart_model_summary.csv")
print("PCA info:", pca_info)
```
